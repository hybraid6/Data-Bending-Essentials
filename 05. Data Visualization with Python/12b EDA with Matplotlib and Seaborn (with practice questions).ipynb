## Exploratory Data Analysis (EDA) with Matplotlib and Seaborn

### What Is EDA?

- Exploratory Data Analysis (EDA) is the process of analyzing data sets to summarize their main characteristics,often using visual methods. 
- It helps in understanding the data, finding patterns, detecting anomalies, testing hypotheses, and checking assumptions.

#### Introduction to Matplotlib and Seaborn

- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
- Seaborn is a data visualization library built on top of Matplotlib that provides a high-level interface for drawing attractive statistical graphics.


```python
import warnings
# Suppress FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)  # module="pandas")
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

#### Bar Chart

```python
sns.countplot(x='class', data=titanic, palette='PuBuGn')
```

```python
# Example: Creating a bar chart to show the count of passengers by class
plt.figure(figsize=(8, 6))
sns.countplot(x='class', data=titanic, palette='PuBuGn')
plt.title('Count of Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

#### Histogram

```python
# Example: Creating a histogram to show the distribution of ages
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'].dropna(), bins=20, kde=True, color='#008080')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#### Scatter Plot

```python
# Example: Creating a scatter plot to show the relationship between age and fare
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', hue='class', data=titanic, palette='inferno')
plt.title('Age vs Fare by Class')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Class')
plt.show()
```

#### Box Plot

```python
# Example: Creating a box plot to show the distribution of ages by class
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='age', data=titanic, palette='inferno')
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

#### Pair Plot

```python
# Example: Creating a pair plot to show relationships between multiple variables
sns.pairplot(titanic[['age', 'fare', 'class']], hue='class', palette='inferno')
plt.suptitle('Pair Plot of Age, Fare, and Class', y=1.02)
plt.show()
```
