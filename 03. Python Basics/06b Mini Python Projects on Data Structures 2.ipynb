### Example Solutions

### Solution for Question 1: Task Manager
```
task_list = []

while True:
    print("\nTask Manager Menu:")
    print("1. Add task")
    print("2. Mark task as completed")
    print("3. Remove task")
    print("4. View tasks")
    print("5. Quit")

    choice = input("Choose an option (1-5): ")

    if choice == '1':
        task = input("Enter the task to add: ")
        task_list.append(task)
        print(f"Added task: {task}")
    elif choice == '2':
        task = input("Enter the task to mark as completed: ")
        if task in task_list:
            task_list.remove(task)
            task_list.append(f"{task} (completed)")
            print(f"Marked task as completed: {task}")
        else:
            print(f"Task not found: {task}")
    elif choice == '3':
        task = input("Enter the task to remove: ")
        if task in task_list:
            task_list.remove(task)
            print(f"Removed task: {task}")
        else:
            print(f"Task not found: {task}")
    elif choice == '4':
        print("Current tasks:")
        for idx, task in enumerate(task_list, start=1):
            print(f"{idx}. {task}")
    elif choice == '5':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```

### Solution for Question 2: Contact Book
```
contact_book = {}

while True:
    print("\nContact Book Menu:")
    print("1. Add contact")
    print("2. Remove contact")
    print("3. View contacts")
    print("4. Quit")

    choice = input("Choose an option (1-4): ")

    if choice == '1':
        name = input("Enter the contact name: ")
        phone = input("Enter the contact phone number: ")
        contact_book[name] = phone
        print(f"Added contact: {name}")
    elif choice == '2':
        name = input("Enter the contact name to remove: ")
        if name in contact_book:
            del contact_book[name]
            print(f"Removed contact: {name}")
        else:
            print(f"Contact not found: {name}")
    elif choice == '3':
        print("Current contacts:")
        for name, phone in contact_book.items():
            print(f"Name: {name}, Phone: {phone}")
    elif choice == '4':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```

### Solution for Question 3: Budget Tracker
```
expenses = []

while True:
    print("\nBudget Tracker Menu:")
    print("1. Add expense")
    print("2. Remove expense")
    print("3. View expenses")
    print("4. Quit")

    choice = input("Choose an option (1-4): ")

    if choice == '1':
        description = input("Enter the expense description: ")
        amount = input("Enter the expense amount: ")
        expenses.append((description, float(amount)))
        print(f"Added expense: {description} - ${amount}")
    elif choice == '2':
        description = input("Enter the expense description to remove: ")
        for expense in expenses:
            if expense[0] == description:
                expenses.remove(expense)
                print(f"Removed expense: {description}")
                break
        else:
            print(f"Expense not found: {description}")
    elif choice == '3':
        print("Current expenses:")
        for idx, (description, amount) in enumerate(expenses, start=1):
            print(f"{idx}. {description} - ${amount:.2f}")
    elif choice == '4':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```

### Solution for Question 4: Library System
```
library = []

while True:
    print("\nLibrary System Menu:")
    print("1. Add book")
    print("2. Remove book")
    print("3. View books")
    print("4. Quit")

    choice = input("Choose an option (1-4): ")

    if choice == '1':
        title = input("Enter the book title: ")
        author = input("Enter the book author: ")
        library.append((title, author))
        print(f"Added book: {title} by {author}")
    elif choice == '2':
        title = input("Enter the book title to remove: ")
        for book in library:
            if book[0] == title:
                library.remove(book)
                print(f"Removed book: {title}")
                break
        else:
            print(f"Book

 not found: {title}")
    elif choice == '3':
        print("Current books:")
        for idx, (title, author) in enumerate(library, start=1):
            print(f"{idx}. {title} by {author}")
    elif choice == '4':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```

### Solution for Question 5: Movie Collection
```
movies = []

while True:
    print("\nMovie Collection Menu:")
    print("1. Add movie")
    print("2. Remove movie")
    print("3. View movies")
    print("4. Quit")

    choice = input("Choose an option (1-4): ")

    if choice == '1':
        title = input("Enter the movie title: ")
        director = input("Enter the movie director: ")
        movies.append((title, director))
        print(f"Added movie: {title} directed by {director}")
    elif choice == '2':
        title = input("Enter the movie title to remove: ")
        for movie in movies:
            if movie[0] == title:
                movies.remove(movie)
                print(f"Removed movie: {title}")
                break
        else:
            print(f"Movie not found: {title}")
    elif choice == '3':
        print("Current movies:")
        for idx, (title, director) in enumerate(movies, start=1):
            print(f"{idx}. {title} directed by {director}")
    elif choice == '4':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```
