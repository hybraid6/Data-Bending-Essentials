### Project 1: Number Guessing Game
Create a program where the computer randomly selects a number between 1 and 100. The user has to guess the number, and the program provides feedback whether the guess is too high, too low, or correct.

```
import random

def number_guessing_game():
    number_to_guess = random.randint(1, 100)  # Computer randomly selects a number between 1 and 100
    attempts = 0

    while True:
        guess = input("Guess the number (between 1 and 100): ")  # User inputs a guess

        if guess.isdigit():  # Check if the input is a valid number
            guess = int(guess)
            attempts += 1
            if guess < number_to_guess:
                print("Too low!")
            elif guess > number_to_guess:
                print("Too high!")
            else:
                print(f"Congratulations! You guessed the number in {attempts} attempts.")
                break
        else:
            print("Please enter a valid number.")

number_guessing_game()
```
**Explanation**:
- The program generates a random number between 1 and 100.
- The user is prompted to guess the number.
- The program checks if the guess is too high, too low, or correct.
- If the guess is correct, the program congratulates the user and ends. If not, it provides feedback and continues.

### Project 2: Simple To-Do List
Create a program that allows users to add, remove, and view tasks in a to-do list.

```
def todo_list():
    tasks = []  # Initialize an empty list to store tasks

    while True:
        print("\n1. Add task")
        print("2. Remove task")
        print("3. View tasks")
        print("4. Quit")
        
        choice = input("Choose an option: ")

        if choice == '1':
            task = input("Enter a task to add: ")
            tasks.append(task)  # Add the task to the list
            print("Task added.")
        elif choice == '2':
            task = input("Enter the task to remove: ")
            if task in tasks:
                tasks.remove(task)  # Remove the task from the list
                print("Task removed.")
            else:
                print("Task not found.")
        elif choice == '3':
            print("\nTo-Do List:")
            for idx, task in enumerate(tasks, 1):
                print(f"{idx}. {task}")  # Display each task with an index
        elif choice == '4':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

todo_list()
```
**Explanation**:
- The program maintains a list of tasks.
- The user can add tasks, remove tasks, and view the current list of tasks.
- The menu is displayed in a loop, allowing the user to perform multiple actions until they choose to quit.

### Project 3: Basic Unit Converter
Create a program that converts units of length (e.g., inches to centimeters, meters to feet).

```
def unit_converter():
    while True:
        print("\nUnit Converter")
        print("1. Inches to Centimeters")
        print("2. Centimeters to Inches")
        print("3. Meters to Feet")
        print("4. Feet to Meters")
        print("5. Quit")
        
        choice = input("Choose an option: ")

        if choice == '1':
            inches = float(input("Enter inches: "))
            centimeters = inches * 2.54  # Conversion factor for inches to centimeters
            print(f"{inches} inches is {centimeters:.2f} centimeters.")
        elif choice == '2':
            centimeters = float(input("Enter centimeters: "))
            inches = centimeters / 2.54  # Conversion factor for centimeters to inches
            print(f"{centimeters} centimeters is {inches:.2f} inches.")
        elif choice == '3':
            meters = float(input("Enter meters: "))
            feet = meters * 3.28084  # Conversion factor for meters to feet
            print(f"{meters} meters is {feet:.2f} feet.")
        elif choice == '4':
            feet = float(input("Enter feet: "))
            meters = feet / 3.28084  # Conversion factor for feet to meters
            print(f"{feet} feet is {meters:.2f} meters.")
        elif choice == '5':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

unit_converter()
```
**Explanation**:
- The program converts between different units of length.
- The user selects the conversion they want to perform from a menu.
- The program performs the conversion and displays the result.
- The loop allows the user to perform multiple conversions until they choose to quit.

### Project 4: Simple BMI Calculator
Create a program that calculates the Body Mass Index (BMI) from user input for weight and height.

```
def bmi_calculator():
    while True:
        weight = input("Enter your weight in kilograms: ")
        height = input("Enter your height in meters: ")

        if weight.replace('.', '', 1).isdigit() and height.replace('.', '', 1).isdigit():
            weight = float(weight)
            height = float(height)
            bmi = weight / (height ** 2)  # BMI formula
            print(f"Your BMI is: {bmi:.2f}")
            break
        else:
            print("Please enter valid numbers for weight and height.")

bmi_calculator()
```
**Explanation**:
- The program calculates BMI using the formula: BMI = weight / (height * height).
- The user inputs their weight and height.
- The program checks if the inputs are valid numbers, performs the calculation, and displays the BMI.

### Project 5: Temperature Converter
Create a program that converts temperatures between Celsius, Fahrenheit, and Kelvin.

```
def temperature_converter():
    while True:
        print("\nTemperature Converter")
        print("1. Celsius to Fahrenheit")
        print("2. Fahrenheit to Celsius")
        print("3. Celsius to Kelvin")
        print("4. Kelvin to Celsius")
        print("5. Quit")
        
        choice = input("Choose an option: ")

        if choice == '1':
            celsius = float(input("Enter temperature in Celsius: "))
            fahrenheit = (celsius * 9/5) + 32  # Conversion formula for Celsius to Fahrenheit
            print(f"{celsius}°C is {fahrenheit:.2f}°F.")
        elif choice == '2':
            fahrenheit = float(input("Enter temperature in Fahrenheit: "))
            celsius = (fahrenheit - 32) * 5/9  # Conversion formula for Fahrenheit to Celsius
            print(f"{fahrenheit}°F is {celsius:.2f}°C.")
        elif choice == '3':
            celsius = float(input("Enter temperature in Celsius: "))
            kelvin = celsius + 273.15  # Conversion formula for Celsius to Kelvin
            print(f"{celsius}°C is {kelvin:.2f}K.")
        elif choice == '4':
            kelvin = float(input("Enter temperature in Kelvin: "))
            celsius = kelvin - 273.15  # Conversion formula for Kelvin to Celsius
            print(f"{kelvin}K is {celsius:.2f}°C.")
        elif choice == '5':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

temperature_converter()
```
**Explanation**:
- The program converts temperatures between Celsius, Fahrenheit, and Kelvin.
- The user selects the conversion they want to perform from a menu.
- The program performs the conversion using the appropriate formula and displays the result.
- The loop allows the user to perform multiple conversions until they choose to quit.
